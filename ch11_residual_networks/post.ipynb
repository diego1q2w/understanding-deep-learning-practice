```python
from config import ExperimentConfig
from trainer_utils import create_and_run_trainner
import matplotlib.pyplot as plt
```

```python
config = ExperimentConfig.load_with_overrides(
    'configs/base.yaml',
    'configs/post_activation.yaml'
)
```

```python
results = create_and_run_trainner(config)
```

```python
fig, axes = plt.subplots(2, 2, figsize=(12, 10))

# Loss
axes[0, 0].plot(results['train_losses'], label='Train')
axes[0, 0].plot(results['val_losses'], label='Val')
axes[0, 0].set_title('Loss')
axes[0, 0].set_xlabel('Epoch')
axes[0, 0].legend()

# Accuracy
axes[0, 1].plot(results['train_accs'], label='Train')
axes[0, 1].plot(results['val_accs'], label='Val')
axes[0, 1].set_title('Accuracy')
axes[0, 1].set_xlabel('Epoch')
axes[0, 1].legend()

# Gradient magnitudes over time
if results['gradient_history'] is not None:
    im = axes[1, 0].imshow(results['gradient_history'].T, aspect='auto', cmap='viridis')
    axes[1, 0].set_title('Gradient Magnitudes')
    axes[1, 0].set_xlabel('Epoch')
    axes[1, 0].set_ylabel('Conv Layer Index')
    plt.colorbar(im, ax=axes[1, 0])

# Epoch times
axes[1, 1].plot(results['epoch_times'])
axes[1, 1].set_title('Epoch Time')
axes[1, 1].set_xlabel('Epoch')
axes[1, 1].set_ylabel('Time (s)')

plt.tight_layout()
plt.show()
```

```python
import numpy as np
gradient_history = results['gradient_history']

if gradient_history is not None:
    fig, axes = plt.subplots(1, 2, figsize=(16, 6))

    # 1. Heatmap: Epochs vs Layers
    im = axes[0].imshow(gradient_history.T, aspect='auto', cmap='viridis',
                        interpolation='nearest', origin='lower')
    axes[0].set_xlabel('Epoch', fontsize=12)
    axes[0].set_ylabel('Conv Layer Index', fontsize=12)
    axes[0].set_title(f'{config.model.net_type.upper()}-Activation\nGradient Magnitude Heatmap',
                      fontsize=14, fontweight='bold')
    cbar = plt.colorbar(im, ax=axes[0])
    cbar.set_label('Normalized Gradient Magnitude', fontsize=10)

    # 2. Line plot: Each layer over time
    num_layers = gradient_history.shape[1]
    colors = plt.cm.rainbow(np.linspace(0, 1, num_layers))

    for layer_idx in range(num_layers):
        axes[1].plot(gradient_history[:, layer_idx],
                    label=f'Layer {layer_idx}' if layer_idx % 3 == 0 else '',
                    color=colors[layer_idx], alpha=0.7, linewidth=1.5)

    axes[1].set_xlabel('Epoch', fontsize=12)
    axes[1].set_ylabel('Gradient Magnitude', fontsize=12)
    axes[1].set_title(f'{config.model.net_type.upper()}-Activation\nGradient Flow Over Training',
                      fontsize=14, fontweight='bold')
    axes[1].set_yscale('log')  # Log scale to see small gradients
    axes[1].grid(True, alpha=0.3)
    axes[1].legend(loc='upper right', fontsize=8, ncol=2)

    plt.tight_layout()
    plt.savefig(f'gradients_{config.model.net_type}.png', dpi=150, bbox_inches='tight')
    plt.show()
else:
    print("⚠️ No gradient history available!")
```
