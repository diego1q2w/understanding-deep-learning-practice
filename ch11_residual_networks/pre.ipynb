```python
from config import ExperimentConfig
from trainer_utils import create_and_run_trainner
import matplotlib.pyplot as plt
```

```python
config = ExperimentConfig.load_with_overrides(
    'configs/base.yaml',
    'configs/pre_activation.yaml'
)
```

```python
results = create_and_run_trainner(config)
```

```python
fig, axes = plt.subplots(2, 2, figsize=(12, 10))

# Loss
axes[0, 0].plot(results['train_losses'], label='Train')
axes[0, 0].plot(results['val_losses'], label='Val')
axes[0, 0].set_title('Loss')
axes[0, 0].set_xlabel('Epoch')
axes[0, 0].legend()

# Accuracy
axes[0, 1].plot(results['train_accs'], label='Train')
axes[0, 1].plot(results['val_accs'], label='Val')
axes[0, 1].set_title('Accuracy')
axes[0, 1].set_xlabel('Epoch')
axes[0, 1].legend()

# Gradient magnitudes over time
if results['gradient_history'] is not None:
    im = axes[1, 0].imshow(results['gradient_history'].T, aspect='auto', cmap='viridis')
    axes[1, 0].set_title('Gradient Magnitudes')
    axes[1, 0].set_xlabel('Epoch')
    axes[1, 0].set_ylabel('Conv Layer Index')
    plt.colorbar(im, ax=axes[1, 0])

# Epoch times
axes[1, 1].plot(results['epoch_times'])
axes[1, 1].set_title('Epoch Time')
axes[1, 1].set_xlabel('Epoch')
axes[1, 1].set_ylabel('Time (s)')

plt.tight_layout()
plt.show()
```
